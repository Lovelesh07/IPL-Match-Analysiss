**2. Data Cleaning**
○ Check for null values and handle them appropriately.
○ Correct column names if needed (e.g., team names or venue names with
inconsistent formatting).
○ Drop irrelevant columns (if any) after justification.

```python
deliveries.isnull().sum()
```

```python
matches.isnull().sum()
```

```python
deliveries['extra_runs'].fillna(0, inplace=True)
deliveries['is_wicket'] = deliveries['player_dismissed'].notna().astype(int)
```

```python
most_frequent_city = matches['city'].mode()[0]
matches['city'].fillna(most_frequent_city, inplace=True)
for col in ['winner', 'player_of_match', 'result', 'umpire1', 'umpire2']:
     matches[col].fillna('Not Available', inplace=True)

if 'umpire3' in matches.columns:
  matches.drop('umpire3', axis=1, inplace=True)
```

```python
matches['team1'] = matches['team1'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
matches['team2'] = matches['team2'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
matches['winner'] = matches['winner'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
matches['toss_winner'] = matches['toss_winner'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
matches['venue'] = matches['venue'].replace('Rajiv Gandhi International Stadium, Uppal', 'Rajiv Gandhi International Stadium')

team_name_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Pune Warriors': 'Rising Pune Supergiants',
    'Rising Pune Supergiants': 'Rising Pune Supergiants',
    'Deccan Chargers': 'Sunrisers Hyderabad',
}
matches['team1'] = matches['team1'].replace(team_name_mapping)
matches['team2'] = matches['team2'].replace(team_name_mapping)
matches['winner'] = matches['winner'].replace(team_name_mapping)
matches['toss_winner'] = matches['toss_winner'].replace(team_name_mapping)

venue_name_mapping = {
    'M Chinnaswamy Stadium': 'M.Chinnaswamy Stadium',
    'Punjab Cricket Association Stadium, Mohali': 'Punjab Cricket Association Stadium',
    'Punjab Cricket Association Stadium, Mohali': 'Punjab Cricket Association Stadium',
    'Wankhede Stadium, Mumbai': 'Wankhede Stadium',
}
matches['venue'] = matches['venue'].replace(venue_name_mapping)

print("Unique Team Names:", matches['team1'].unique())
print("\nUnique Venue Names:", matches['venue'].unique())
```

```python
deliveries['batting_team'] = deliveries['batting_team'].replace(team_name_mapping)
deliveries['bowling_team'] = deliveries['bowling_team'].replace(team_name_mapping)

print("Unique Batting Teams:", deliveries['batting_team'].unique())
print("\nUnique Bowling Teams:", deliveries['bowling_team'].unique())
```

```python
matches.drop(['umpire1', 'umpire2', 'dl_applied', 'id'], axis=1, inplace=True)

deliveries.drop(['dismissal_kind', 'fielder', 'player_dismissed'], axis=1, inplace=True)

print("Matches DataFrame:")
display(matches.head())
print("\nDeliveries DataFrame:")
display(deliveries.head())
```

**3. Exploratory Data Analysis (EDA):**
Use appropriate functions to answer the following:
Match-Level Analysis (Using Matches.csv):

Q1: Which team won the most matches in the dataset?


```python
most_wins = matches['winner'].value_counts().idxmax()
print(f"The team that won the most matches is: {most_wins}")
```

Q2: What is the average winning margin (runs and wickets)?

```python
avg_win_by_runs = matches['win_by_runs'].mean()
avg_win_by_wickets = matches['win_by_wickets'].mean()

print(f"Average winning margin by runs: {avg_win_by_runs:.2f}")
print(f"Average winning margin by wickets: {avg_win_by_wickets:.2f}")
```

Q3: What are the top 5 cities where matches were held?

```python
top_5_cities = matches['city'].value_counts().head(5)
print("Top 5 cities where matches were held:\n", top_5_cities)
```

Q4: Find the venue with the most matches hosted

```python
most_matches_venue = matches['venue'].value_counts().idxmax()
print(f"The venue with the most matches hosted is: {most_matches_venue}")
```

Q5: Which player won the most "Player of the Match" awards?

```python
most_player_of_match = matches['player_of_match'].value_counts().idxmax()
print(f"The player who won the most 'Player of the Match' awards is: {most_player_of_match}")
```

**4. Ball-Level Analysis (Using Deliveries.csv):**
○ Q6: Which batsman scored the most runs overall?


```python
most_runs_batsman = deliveries.groupby('batsman')['batsman_runs'].sum().idxmax()
print(f"The batsman who scored the most runs overall is: {most_runs_batsman}")
```

Q7: Which bowler took the most wickets?

```python
most_wickets_bowler = deliveries[deliveries['is_wicket'] == 1]['bowler'].value_counts().idxmax()
print(f"The bowler who took the most wickets is: {most_wickets_bowler}")
```

Q8: What is the distribution of extras (wide, no-ball, leg-byes)?

```python
wide_runs_total = deliveries['wide_runs'].sum()
noball_runs_total = deliveries['noball_runs'].sum()
legbye_runs_total = deliveries['legbye_runs'].sum()
bye_runs_total = deliveries['bye_runs'].sum()

print(f"Total wide runs: {wide_runs_total}")
print(f"Total no-ball runs: {noball_runs_total}")
print(f"Total leg-bye runs: {legbye_runs_total}")
print(f"Total bye runs: {bye_runs_total}")

total_extras = wide_runs_total + noball_runs_total + legbye_runs_total + bye_runs_total

print(f"\nPercentage distribution of extras:")
print(f"Wide runs: {wide_runs_total / total_extras * 100:.2f}%")
print(f"No-ball runs: {noball_runs_total / total_extras * 100:.2f}%")
print(f"Leg-bye runs: {legbye_runs_total / total_extras * 100:.2f}%")
print(f"Bye runs: {bye_runs_total / total_extras * 100:.2f}%")
```

Q9: Which team scored the highest runs in a single match?

```python
highest_runs_team = deliveries.groupby(['match_id', 'batting_team'])['total_runs'].sum().idxmax()
print(f"the highest runs in a single match is: {highest_runs_team[1]} in match {highest_runs_team[0]}")
```

Q10: Plot the trend of total runs scored per over in a match
(visualization)

```python
import matplotlib.pyplot as plt

runs_per_over = deliveries.groupby(['match_id', 'over'])['total_runs'].sum().reset_index()

plt.figure(figsize=(12, 6))
for match_id in runs_per_over['match_id'].unique():
    match_data = runs_per_over[runs_per_over['match_id'] == match_id]
    plt.plot(match_data['over'], match_data['total_runs'], label=f'Match {match_id}')

plt.xlabel('Over')
plt.ylabel('Total Runs Scored')
plt.title('Trend of Total Runs Scored per Over in a Match')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

**5. Visualization: Use Matplotlib or Seaborn to create the following visualizations:**

○ Plot the top 5 teams with the most wins.


```python
import matplotlib.pyplot as plt

top_5_teams = matches['winner'].value_counts().head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_5_teams.index, top_5_teams.values)
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Top 5 Teams with the Most Wins')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

○ Bar chart of the top 5 batsmen with the highest runs.

```python
top_5_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_5_batsmen.index, top_5_batsmen.values)
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.title('Top 5 Batsmen with the Highest Runs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

○ Distribution of winning margins (runs and wickets) using histograms

```python
plt.figure(figsize=(10, 6))
plt.hist(matches['win_by_runs'], bins=20, edgecolor='black')
plt.xlabel('Winning Margin (Runs)')
plt.ylabel('Frequency')
plt.title('Distribution of Winning Margins (Runs)')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(matches['win_by_wickets'], bins=10, edgecolor='black')
plt.xlabel('Winning Margin (Wickets)')
plt.ylabel('Frequency')
plt.title('Distribution of Winning Margins (Wickets)')
plt.show()
```

○ Line plot showing runs scored across overs in a specific match

```python
match_id = 1

match_data = deliveries[deliveries['match_id'] == match_id]
runs_per_over = match_data.groupby('over')['total_runs'].sum()

plt.figure(figsize=(10, 6))
plt.plot(runs_per_over.index, runs_per_over.values)
plt.xlabel('Over')
plt.ylabel('Runs Scored')
plt.title(f'Runs Scored Across Overs in Match {match_id}')
plt.grid(True)
plt.show()
```
